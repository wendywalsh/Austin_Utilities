```python
import pandas as pd
import numpy as np
import os


```

```python
population = pd.read_csv('../austin_population2001-2016.csv')
population
```

```python
# read in CSV from austin open data
data = pd.read_csv('austin_waste_and_diversion.csv')
data
```

```python
# List of unique drop off sites. Many are missed type drop off. 
data.dropoff_site.unique()
```

```python
# List of load type.
data.load_type.unique()
```

```python
# identify number of row for "sweeping load type. Most have Nan.
data[data.load_type == 'SWEEPING'].head()
```

```python
# remove "sweeping" load type and create new csv.
no_sweeping_included = data[data.load_type != 'SWEEPING']
```

```python
#Reformat time to allow group bys by year, month and add to DF
load_time_formatted = pd.to_datetime(no_sweeping_included['load_time'], format='%m/%d/%y %H:%M')
no_sweeping_included['load_time_formatted'] = load_time_formatted.values
```

```python
# create new year and month column for group by options
no_sweeping_included['year'] = no_sweeping_included['load_time_formatted'].dt.year
no_sweeping_included['month'] = no_sweeping_included['load_time_formatted'].dt.month

no_sweeping_included.to_csv('new_no_sweeping_included_waste_diversion.csv')
no_sweeping_included
```

### total load weight by year

```python
# Group by year and sum total weight
total_load_by_year = pd.DataFrame(
    no_sweeping_included.groupby(['year'])['load_weight'].sum())
total_load_by_year = total_load_by_year.reset_index()

# Rename year to Year to match 'population' Year column
total_load_by_year = total_load_by_year.rename(columns={"year": "Year", "load_weight":"total_load"})
total_load_by_year
```

#### combined population with total load weight

```python
# merge total_load_by_year and population DF
combined_population_loadweight = pd.merge(total_load_by_year, population, how='inner', on='Year', left_on=None, right_on=None,
                                          left_index=False, right_index=False, sort=True,
                                          suffixes=('_x', '_y'), copy=True, indicator=False)
```

## Per capita total load weight by year

```python
# Dropped 2001-2003 and 2017 as only partial year data provided
combined_population_loadweight = combined_population_loadweight.iloc[1:]

# load_weight / total_population for load per capita
combined_population_loadweight['load_per_capita'] = combined_population_loadweight['total_load'] / \
    combined_population_loadweight['total_area_population']
combined_population_total_loadweight = combined_population_loadweight.rename(columns={"load_per_capita": "total_load_per_capita"})
combined_population_total_loadweight
```

#### Per Capita Lanfill only weight by year (all recycling load filtered out)

```python
# garbage weight only by filtering out any load type having to do with recycling, bulk pickup,yard waste
landfill_only = no_sweeping_included[~no_sweeping_included['load_type'].isin(['RECYCLING - SINGLE STREAM', 'YARD TRIMMING', 'RECYCLING - COMINGLE', 'RECYCLING - PAPER',
                                                                              'BRUSH', 'MULCH', 'RECYCLED METAL', 'TIRES', 'RECYCLING - PLASTIC BAGS', 'MATTRESS', 'YARD TRIMMING - X-MAS TREES'
                                                                              'RECYCLING - COMINGLE', ])]
landfill_only.head()
```

```python
# Group by year and sum total weight
landfill_only = pd.DataFrame(
    landfill_only.groupby(['year'])['load_weight'].sum())
landfill_only = landfill_only.reset_index()
```

```python
# Rename year to Year to match 'population' Year column
landfill_only = landfill_only.rename(columns={"year": "Year", "load_weight":"landfill_only_load"})
landfill_only

# merge total_load_by_year and population DF
combined_population_landfill_only = pd.merge(landfill_only, population, how='inner', on='Year', left_on=None, right_on=None,
                                             left_index=False, right_index=False, sort=True,
                                             suffixes=('_x', '_y'), copy=True, indicator=False)

# per capita load landfill only

combined_population_landfill_only['load_per_capita'] = combined_population_landfill_only['landfill_only_load'] / \
    combined_population_landfill_only['total_area_population']
combined_population_landfill_only = combined_population_landfill_only.rename(columns={"load_per_capita": "landfill_only_load_per_capita"})
combined_population_landfill_only
```

### recycling waste by filtering load type. 

```python
# Recycled weight only by filtering load types to exclude garbage collect, dead animals, litter
recycle_only = no_sweeping_included[~no_sweeping_included['load_type'].isin(
    ['GARBAGE COLLECTIONS', 'BAGGED LITTER', 'DEAD ANIMAL', 'LITTER', 'MIXED LITTER','BULK'])]
recycle_only.head()
```

### Per Capita recycling load weight


```python
# Group by year and sum total weight
recycle_only = pd.DataFrame(recycle_only.groupby(['year'])['load_weight'].sum())
recycle_only = recycle_only.reset_index()


# Rename year to Year to match 'population' Year column
recycle_only = recycle_only.rename(columns={"year": "Year"})
recycle_only

# merge total_load_by_year and population DF
combined_population_recycle_only = pd.merge(recycle_only, population, how='inner', on='Year', left_on=None, right_on=None,
                                            left_index=False, right_index=False, sort=True,
                                            suffixes=('_x', '_y'), copy=True, indicator=False)
combined_population_recycle_only

# per capita load landfill only

combined_population_recycle_only['load_per_capita'] = combined_population_recycle_only["load_weight"] / \
    combined_population_recycle_only['total_area_population']
combined_population_recycle_only

combined_population_recycle_only = combined_population_recycle_only.rename(columns={"load_weight":"recycle_only_load", "load_per_capita":"recycle_only_load_per_capita"})
```

```python
# merge total load and landfill only
combined_total_and_landfill = pd.merge(combined_population_total_loadweight,combined_population_landfill_only, how='inner', on='Year', left_on=None, right_on=None,
                                             left_index=False, right_index=False, sort=True,
                                             suffixes=('_x', '_y'), copy=True, indicator=False)
```

```python
# combined total load, landfill only, and recycle only
all_by_year = pd.merge(combined_total_and_landfill, combined_population_recycle_only, how='inner', on='Year', left_on=None, right_on=None,
                                             left_index=False, right_index=False, sort=True,
                                             suffixes=('_x', '_y'), copy=True, indicator=False)
all_by_year.head()
```

```python
all_by_year = all_by_year[['Year','total_area_population','annualized_growth_rate','total_load','total_load_per_capita','landfill_only_load','landfill_only_load_per_capita','recycle_only_load','recycle_only_load_per_capita']]

all_by_year
```

#### Percent recycle/reuse/recovered diverted by landfill

```python
all_by_year['%oftotal_load_recycled'] = all_by_year['recycle_only_load']/all_by_year['total_load']

all_by_year.to_csv('new_waste_analysis_by_year.csv')
```
