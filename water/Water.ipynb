```python
import pandas as pd
import numpy as np
import os
```

```python
commercial_water_consumption = pd.read_csv('Austin_Water_-_Commercial_Water_Consumption.csv')
# reformat date. Day of date is not part of original data. 
commercial_water_consumption['YearMonth'] = pd.to_datetime(commercial_water_consumption['YearMonth'], format='%Y%m')
commercial_water_consumption.head()
```

```python
residential_water_consumption = pd.read_csv('Austin_Water_-_Residential_Water_Consumption.csv')
# reformat date. Day of date is not part of original data. 
residential_water_consumption['YearMonth'] = pd.to_datetime(residential_water_consumption['YearMonth'], format='%Y%m')
residential_water_consumption.head()
```

```python
# Union commercial and residential data
dataframes = [commercial_water_consumption, residential_water_consumption]
compiled_resident_commercial = pd.concat(dataframes)

# reformat date. Day of date is not part of original data. 
compiled_resident_commercial['YearMonth'] = pd.to_datetime(compiled_resident_commercial['YearMonth'], format='%Y%m')

#create csv
compiled_resident_commercial.to_csv('new_compiled_resident_commercial.csv')

compiled_resident_commercial.head()
```
